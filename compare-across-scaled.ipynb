```python
lowres = pd.read_csv('SRR606249.100k.gather.csv').set_index('match_name')
medres = pd.read_csv('SRR606249.10k.gather.csv').set_index('match_name')
highres = pd.read_csv('SRR606249.1k.gather.csv').set_index('match_name')
```

```python
len(lowres)
```

```python
len(medres)
```

```python
len(highres)
```

```python
merged = lowres.join(highres, on='match_name', rsuffix='_high').join(medres, on='match_name', rsuffix='_med')
```

```python
plt.plot(merged.f_unique_to_query, merged.f_unique_to_query_high, '*')
plt.plot([0, .02], [0, .02], '--')
plt.xlabel('high res matches (scaled=1k)')
plt.ylabel('low res matches (scaled=100k)')
plt.savefig('SRR606249.1k-vs-100k.png')
```

```python
plt.plot(merged.f_unique_to_query, merged.f_unique_to_query_med, '*')
plt.plot([0, .02], [0, .02], '--')
plt.xlabel('high res matches (scaled=1k)')
plt.ylabel('medres matches (scaled=10k)')
plt.title('SRR606249: gather 1k vs 10k scaled')
plt.savefig('SRR606249.1k-vs-10k.png')
```
